```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndimage
import pandas as pd
```

```python
def graficas(df,title, low_lim,sup_lim):
    new_columns = ['Longitud de Onda', 'Flujo']  # Replace with your desired column names
    # Assign new column names to the DataFrame
    df.columns = new_columns
    df = df.iloc[1:]
    x=df["Longitud de Onda"]
    x=list(x)
    x = [float(value) for value in x]
    y=df["Flujo"]
    y=list(y)
    y = [float(value) for value in y]
    starting_index = np.argmin(np.abs(np.array(x) - low_lim))
    superior_limit_index = np.argmin(np.abs(np.array(x) - sup_lim))
    plt.plot(x,y)
    plt.xlabel("Longitud de Onda (A)")
    plt.ylabel("Flujo")
    plt.xlim(x[starting_index], x[superior_limit_index])
    plt.title(title)
    plt.grid(True)
    plt.show()
    return df
    
```

```python
def grafica2(df, title, low_lim, sup_lim):
    new_columns = ['Longitud de Onda', 'Flujo']  # Replace with your desired column names
    # Assign new column names to the DataFrame
    df.columns = new_columns
    df = df.iloc[1:]
    x = df["Longitud de Onda"].astype(float)
    y = df["Flujo"].astype(float)

    # Determine the indices for slicing the DataFrame
    starting_index = np.argmin(np.abs(x - low_lim))
    superior_limit_index = np.argmin(np.abs(x - sup_lim))

    # Create a new DataFrame with data within the specified limits
    sliced_df = df[starting_index:superior_limit_index + 1]

    # Plotting
    plt.plot(x, y)
    plt.xlabel("Longitud de Onda (A)")
    plt.ylabel("Flujo")
    plt.xlim(x[starting_index], x[superior_limit_index])
    plt.title(title)
    plt.grid(True)
    plt.show()

    return sliced_df

# Example usage:
# sliced_dataframe = graficas(df, "Title", low_lim, sup_lim)
```

```python
from scipy.optimize import curve_fit
```

```python
def generate_parabola(x,h,k,a):
    return (a*((x-h)**2)) + k
```

```python
Hierros=[]
```

```python
INF=6136.5
SUP=6136.7

Fe1=grafica2(procyon,"Procyon 6100A-6300A",INF,SUP)

X1= list(Fe1["Longitud de Onda"])
X1 = [float(element) for element in X1]
X1=np.array(X1)

Y=list(Fe1["Flujo"])
Y = [float(element) for element in Y]
Y=np.array(Y)

params, covariance = curve_fit(generate_parabola, X1, Y,p0=[6136.6,1000,1.5])
plt.scatter(X1,Y)
plt.plot(X1, generate_parabola(X1,*params), color='red', label='Fitted Curve')
centro=params[0]
plt.show()
print(centro)
Hierros.append(centro)
```
