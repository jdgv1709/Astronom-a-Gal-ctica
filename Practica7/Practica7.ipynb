```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as ndimage
import pandas as pd
```

```python
def graficas(df,title, low_lim,sup_lim):
    new_columns = ['Longitud de Onda', 'Flujo']  # Replace with your desired column names
    # Assign new column names to the DataFrame
    df.columns = new_columns
    df = df.iloc[1:]
    x=df["Longitud de Onda"]
    x=list(x)
    x = [float(value) for value in x]
    y=df["Flujo"]
    y=list(y)
    y = [float(value) for value in y]
    starting_index = np.argmin(np.abs(np.array(x) - low_lim))
    superior_limit_index = np.argmin(np.abs(np.array(x) - sup_lim))
    plt.plot(x,y)
    plt.xlabel("Longitud de Onda (A)")
    plt.ylabel("Flujo")
    plt.xlim(x[starting_index], x[superior_limit_index])
    plt.title(title)
    plt.grid(True)
    plt.show()
    
```
