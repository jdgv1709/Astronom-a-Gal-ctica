```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\juang\Documents\Tareas JD\Universidad\Materias\7 Semestre\Astronomia Galactica\Practica 11\Variables Magallanes.csv")
df.shape
```

```python
df=df.sort_values(by="P")
```

```python
df.sort_values(by="magMin")
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
```

```python
plt.scatter(df["P"],df["magMax"], marker=".")
#plt.scatter(df["P"],df["magMin"])
plt.xscale('log')
plt.gca().invert_yaxis()
plt.title("Magnitudes Superiores de Estrellas Cefeidas de SMC")
slope, intercept, r_value, p_value, std_err = linregress(np.log(df["P"]),df["magMax"])
plt.plot(df["P"], slope*np.log(df["P"]) + intercept, color='red', label="P* "+str(slope)+" + "+str(intercept))
plt.legend()
plt.xlabel("Periodo [Dias]")
plt.ylabel("Magnitud")
```

```python
#plt.scatter(df["P"],df["magMax"])
plt.scatter(df["P"],df["magMin"], marker=".")
plt.xscale('log')
plt.gca().invert_yaxis()
slope, intercept, r_value, p_value, std_err = linregress(np.log(df["P"]),df["magMin"])
plt.plot(df["P"], slope*np.log(df["P"]) + intercept, color='red', label="P* "+str(slope)+" + "+str(intercept))
plt.legend()
plt.title("Magnitudes Inferiores de Estrellas Cefeidas de SMC")
plt.xlabel("Periodo [Dias]")
plt.ylabel("Magnitud")
```

```python
plt.scatter(df["P"],df["magMax"], label="Maximos", marker=".")
#plt.scatter(df["P"],df["magMin"])
slope, intercept, r_value, p_value, std_err = linregress(np.log(df["P"]),df["magMax"])
plt.plot(df["P"], slope*np.log(df["P"]) + intercept, color='red')


#plt.scatter(df["P"],df["magMax"])
plt.scatter(df["P"],df["magMin"], label="Minimos", marker="x")
plt.xscale('log')
plt.gca().invert_yaxis()
slope, intercept, r_value, p_value, std_err = linregress(np.log(df["P"]),df["magMin"])
plt.plot(df["P"], slope*np.log(df["P"]) + intercept, color='red')
plt.legend()
plt.title("Magnitudes  Superiores e Inferiores de Estrellas Cefeidas de SMC")
plt.xlabel("Periodo [Dias]")
plt.ylabel("Magnitud")
```

```python
promedio=(df["magMax"]+df["magMin"])/2
plt.scatter(df["P"],promedio, marker=".")
#plt.scatter(df["P"],df["magMin"])
plt.xscale('log')
plt.gca().invert_yaxis()
slope, intercept, r_value, p_value, std_err = linregress(np.log(df["P"]),promedio)
plt.plot(df["P"], slope*np.log(df["P"]) + intercept, color='red', label="P* "+str(slope)+" + "+str(intercept))
plt.legend()
plt.title("Magnitudes Promedio de Estrellas Cefeidas de SMC")
plt.xlabel("Periodo [Dias]")
plt.ylabel("Magnitud")
```

```python
dif=abs(df["magMax"]-df["magMin"])
plt.scatter(df["P"],dif)
#plt.scatter(df["P"],df["magMin"])
plt.xscale('log')
#plt.gca().invert_yaxis()
plt.title(" Diferencia de Magnitudes de Estrellas Cefeidas de SMC")
plt.xlabel("Periodo [Dias]")
plt.ylabel("Diferencia Absoluta entre Magnitud Max-Min")
```

```python
mag_abs=-2.81*np.log10(df["P"])-1.43
plt.scatter(df["P"],mag_abs)
plt.xscale('log')
plt.gca().invert_yaxis()
slope, intercept, r_value, p_value, std_err = linregress(np.log(df["P"]),mag_abs)
plt.plot(df["P"], slope*np.log(df["P"]) + intercept, color='red', label="P* "+str(slope)+" + "+str(intercept))
plt.legend()
plt.title("Magnitudes Absoluta Promedio Segun Periodo")
plt.xlabel("Periodo [Dias]")
plt.ylabel("Magnitud Absoluta")
```

```python
dist=10**((mag_abs-promedio-5)/-5)
plt.scatter(df["P"],dist)
plt.xscale('log')
plt.plot(df["P"],np.ones_like(df["P"])*dist.mean(),color="r",label="Distancia Promedio: "+str(dist.mean()/1000)+" Pc")
plt.legend()
plt.xlabel("Periodo [Dias]")
plt.ylabel("Distancia [Pc]")
plt.title("Estimado Distancia")
```

```python
plt.hist(dist)
plt.grid(True)
plt.xlabel("Distancia [Pc]")
```
